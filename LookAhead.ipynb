```python
from wordLists import guesses, answers
from utils import *
```

```python
G = sorted(guesses + answers)
S = sorted(answers)
```

```python
overallBestGuess = None
overallBestScore = 0
for g1 in G:
    splits = getSplits(g1, S, useWords = True)
    worstKey = sorted([(-len(v),k) for k,v in splits.items()])[0][1]
    worstCands = splits[worstKey]
    nextBestGuess = None
    nextBestScore = 0
    for g2 in G:
        splits2 = getSplits(g2, worstCands)
        score = max(splits2.values())
#         print(f"{g1} -> {g2} ==> Worst case {score}")
        if nextBestGuess is None or score <= nextBestScore:
            nextBestScore = score
            nextBestGuess = g2
        
    print(f"Best worst case is {g1} -> {g2} ==> Worst case {nextBestScore}")
    
    if overallBestGuess is None or nextBestScore <= overallBestScore:
        overallBestGuesses = g1
        overallBestScore = nextBestScore
        
        
print(f"Overall best first guess is {overallBestGuess} -> {overallBestScore}")
```
