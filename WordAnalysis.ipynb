```python
from wordfreq import zipf_frequency
from wordLists import *
import pandas as pd
```

```python
import bokeh.plotting as bop
import bokeh.io as boi

boi.output_notebook()
```

```python
allWords = sorted(guesses + answers)
```

```python
df = (
    pd.DataFrame()
    .assign(
        word = allWords,
        freq = [zipf_frequency(x,'en') for x in allWords],
        is_answer = [x in answers for x in allWords]
    )
    .sort_values(by = 'freq')
#     .reset_index()
)
df
```

```python
df[df['is_answer']]
```

```python
p = bop.figure(width = 1000)
answer_df = df[df['is_answer']]
other_df = df[~df['is_answer']]
p.circle(answer_df.index, answer_df['freq'], color='red', size = 2)
p.circle(other_df.index, other_df['freq'], color='blue', size = 2)
bop.show(p)
```
